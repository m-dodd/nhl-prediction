```python
all_passed = True
for c in df1.columns:
    try:
        using = 'test1'
        test = np.isclose(df1[c], df2[c], equal_nan = True).all()
    except:
        using = 'test2'
        test = (df1[c] == df2[c]).all()

    if not test:
        all_passed = False
        print(c, "not equal")

if all_passed:
    print("all columns are equal, hence dataframes are equal")
```
